```python
import numpy as np
import sounddevice as sd
from scipy.io.wavfile import write, read
import matplotlib.pyplot as plt
import librosa.display

sr=44100
```

```python
duration = 5
sr = 44100


time = np.linspace(0, duration, int(sr*duration))
frequency = np.linspace(0, 8000, len(time))
   
output = frequency*time
   
print("this is the output: \n",output)

chirp_wave = np.sin(2 * np.pi * frequency * time)

#sd.play(chirp_wave, samplerate=sr)

sd.wait()
write(f'teamWestCharpPina-chirp.wav', sr, chirp_wave)

# Plot the spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(chirp_wave)), ref=np.max), y_axis='log', x_axis='time', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram of Chirp Tone')
plt.show()
```
